```python
import itertools
```

```python
N = set(['A', 'B', 'C', 'D', 'E', 'F','G', 'H'])
```

```python
powerset_dict = dict()
for l in range(len(N)+1):
    for subset in itertools.combinations(N,l):
        powerset_dict[tuple(sorted(subset))] = [0 for i in N]
```

```python
N_list = list(N)
for i, player in enumerate(N_list):
    for S in itertools.permutations(N):
        S_list = list(S)
        pos = S_list.index(player)
        coalition = tuple(sorted(S_list[:pos]))
        remainder = tuple(sorted(S_list[pos:]))
        powerset_dict[coalition][i]+= 1
   
```
