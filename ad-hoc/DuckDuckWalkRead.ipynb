```python
import os
import pickle
from   matplotlib import pyplot as plt
from   DuckDuckBlipWalk import node
```

```python
def treewalk(node_id=0, indent=0):
    for node in [node for node in results if node.parent_id == node_id]:
        print(node.id, " " * indent, node.text)
        treewalk(node.id, indent + 2)

```

```python
with open(os.path.join(os.getcwd(), os.pardir, 'rory.pkl'), 'rb') as file:
    results = pickle.load(file)
```

```python
rows, cols = 12, 6
fig, ax = plt.subplots(nrows=rows, ncols=cols, sharex=True, sharey=True, figsize=(12,24))
for y in range(rows):
    for x in range(cols):
        ax[y][x].set_axis_off()
        picture = results[y * cols + x].image.resize((240,240))
      
        ax[y][x].imshow(picture)
        ax[y][x].set_title(results[y * cols + x].id)

```
